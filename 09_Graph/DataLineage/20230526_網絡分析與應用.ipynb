## 基礎練習

```python
# 建構圖物件
G = nx.Graph()

# 建立關聯
G.add_edges_from([('Lue', 'Tony'), ('Lue', 'William'), ('Lue', 'Martha'), ('Lue', 'Gordon'),
        ('Qing', 'Paul'), ('Qing', 'Tim'), ('Qing', 'Cora'), ('Qing', 'Ellie'), ('Qing', 'David'),
        ('Helen', 'Bruce'), ('Helen', 'Eric'), 
        ('Job', 'Lue'), ('Job', 'Qing'),('Job', 'Helen')])

# 畫圖
nx.draw(G, with_labels=True, font_size=18)
```

## 中心性評估

```python
# Degree
G.degree()
```

```python
# Closeness
nx.closeness_centrality(G)
```

```python
# Betweenness
nx.betweenness_centrality(G)
```

```python
# Pagerank
nx.pagerank(G)
```

## 資料血緣

```python
df = pd.read_parquet('https://github.com/tlyu0419/DataScience/raw/master/09_Graph/data_lineage_edge_table.parquet')
df
```

```python
DG = nx.DiGraph()
```

```python
gdata = [(source, target, {'endtime': time}) for source, target, time in zip(df['source'], df['target'], df['time'])]
DG.add_edges_from(gdata)
```

```python
print(DG.number_of_nodes())
print(DG.number_of_edges())
```

```python
# tbname = 'table_665'
tbname = 'table_810'
```

```python
nx.single_source_shortest_path_length(DG, tbname, cutoff=99)
```

```python
node_list = list(nx.single_source_shortest_path_length(DG, tbname, cutoff=99))
node_list
```

```python
sdg = nx.subgraph(G = DG, nbunch = node_list)
nx.draw(sdg, with_labels=True, font_size=18)
```

```python
# layout
pos = nx.spring_layout(sdg, k=16, iterations=100, scale =20)

nx.draw(sdg, pos,
        with_labels=True,
        node_color='lightblue',
        arrowsize=40,
        node_size=300,
        alpha=0.7)
```

```python
fig, ax = plt.subplots(figsize=(16, 16))

# sub-graph
node_list = list(nx.single_source_shortest_path_length(DG, tbname, cutoff=4).keys())
sdg = nx.subgraph(G = DG, nbunch = node_list)

# layout
pos = nx.spring_layout(sdg, k=16, iterations=100, scale =20)

nx.draw(sdg, pos,
        with_labels=True,
        node_color='lightblue',
        arrowsize=40,
        node_size=300,
        alpha=0.7,
        ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(16, 16))

# sub-graph
node_list = list(nx.single_source_shortest_path_length(DG, tbname, cutoff=4).keys())
sdg = nx.subgraph(G = DG, nbunch = node_list)

# edgelabel
edge_labels = {}
for k in sdg.edges.keys():
    edge_labels[k] = sdg.edges[k]['endtime']

# layout
pos = nx.spring_layout(sdg, k=16, iterations=100, scale =20)

nx.draw_networkx_edge_labels(sdg, pos,
        edge_labels=edge_labels,
        alpha=0.7,
        ax=ax)

nx.draw(sdg, pos,
        with_labels=True,
        node_color='lightblue',
        arrowsize=40,
        node_size=300,
        alpha=0.7,
        ax=ax)
```
