# 查詢地區號碼

```python
url = 'https://static.104.com.tw/category-tool/json/Area.json'
resp = requests.get(url)
df1 = []
for i in resp.json()[0]['n']:
    ndf = pd.DataFrame(i['n'])
    ndf['city'] = i['des']
    df1.append(ndf)
df1=pd.concat(df1, ignore_index=True)
df1 = df1.loc[:,['city','des','no']]
df1 = df1.sort_values('no')
df1
```

# 職缺清單

## 數據分析

```python
keyword = '數據分析'
jobs = []

for area in df1['no']:
    page = 1
    print('Area:',area)
    print('Job Counts:', len(jobs))
    while True:
        try:            
            url = 'https://www.104.com.tw/jobs/search/list?ro=1&kwop=7&keyword={}&jobcatExpansionType=1&area={}&order=12&asc=0&page={}&mode=s&jobsource=2018indexpoc'.format(keyword, area, page)
            
            resp = requests.get(url,headers=headers)
            for job in resp.json()['data']['list']:
                jobs.append(job['link']['job'])
            if len(resp.json()['data']['list'])<20:
                break
            page += 1
        except:
            print('Retry')

jobs = list(set(jobs))
```

```python
with open('DataAnalystic.pkl', 'wb') as f:
    pickle.dump(jobs, f)
```

# 職缺內容

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36',
           'Referer': 'https://www.104.com.tw/'}
```

## 數據分析

```python
with open('DataAnalystic.pkl', 'rb') as f:
    jobs = pickle.load(f)
```

```python
df = []
for i, job in enumerate(jobs):
    if i%100==0:
        print(i)
    url = 'https:' + job
    jobid = re.findall(r'job/([A-Za-z0-9]{1,})', url)[0]
    # 職缺內容
    url2 = 'https://www.104.com.tw/job/ajax/content/' + jobid
    resp = requests.get(url2, headers=headers)
    # 會有些異常的職缺沒有資料，同 try 函數 parse 資料
    try:
        data = resp.json()['data']
        ndf = pd.DataFrame([{'jobName':data['header']['jobName'], # 職缺名稱
                             'custName':data['header']['custName'], # 公司名稱
                             'custUrl':data['header']['custUrl'], # 公司網址
                             'jobDescription':data['jobDetail']['jobDescription'], # 工作內容
                             'specialty':', '.join([i['description'] for i in data['condition']['specialty']]), # 擅長工具
                             'skill':', '.join([i['description'] for i in data['condition']['skill']]), # 工作技能
                             'other': data['condition']['other'], # 其他條件
                             'workExp': data['condition']['workExp'], # 工作經歷
                             'edu': data['condition']['edu'], # 教育程度
                             'appearDate': data['header']['appearDate'], # 刊登日期
                             'addressRegion':data['jobDetail']['addressRegion'], # 工作地區
                             'addressDetail': data['jobDetail']['addressDetail'], # 工作地址
                             'longitude':data['jobDetail']['longitude'], # 地址經度
                             'latitude':data['jobDetail']['latitude'], # 地址緯度
                             'salary': data['jobDetail']['salary'],
                             'url':url
                            }])
        df.append(ndf)
    except:
        print(url2)
df=pd.concat(df, ignore_index=True)
df
```

```python
df.to_pickle('DataAnalystic.pkl')
```
