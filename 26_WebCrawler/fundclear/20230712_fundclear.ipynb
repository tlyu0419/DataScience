```python
# fundtype
fundtypes = []
for fundtype in soup.find('select', {'name':'fundtype'}).select('option'):
    if fundtype['value']!='':
        fundtypes.append([fundtype['value'], fundtype.text])
# fundtypes
```

```python
yyyymms = ['20201231', '20211231', '20221231', '20230531']
```

```python
urls = []
for agent in agents:
    for fundtype in fundtypes:
        for ym in yyyymms:
            # print(agent, fundtype)
            url = f'https://announce.fundclear.com.tw/MOPSonshoreFundWeb/E02.jsp?fundYm={ym}&agent={agent[0]}&fundtype={fundtype[0]}&curr=&sort=&direction='
            urls.append(url)
len(urls)
```

```python
# 子執行緒的工作函數
def job(url):
    try:
        resp = requests.get(url, headers=header)
        df = pd.read_html(resp.text, header=0)
        df = df[9]
        df = df.loc[df['基金統編'].notnull()]
        df['agent'] = agent[1]
        df['fundtype'] = fundtype[1]
        df['yyyymm'] = ym
    except:
        print(url)

# 建立一個子執行緒
t = threading.Thread(target = job)

# 執行該子執行緒
t.start()

# 主執行緒繼續執行自己的工作
for i in range(3):
  print("Main thread:", i)
  time.sleep(1)

# 等待 t 這個子執行緒結束
t.join()

print("Done.")
```
