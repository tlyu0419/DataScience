```python
def Get_Links(GroupUrl, ToDate):
    urls = []
#     resp = requests.get(url)
    ID = get_groupid(GroupUrl)
    urls += re.findall('url:"(https://www.facebook.com/.*?)/"',resp.text)
    end_cursor = re.findall('end_cursor:"(.*?)",group_id',resp.text)[0]
    utime = datetime.fromtimestamp(int(min(re.findall(r'utime="(.*?)"',resp.text))))
    print('There are {} Links in urls, and lastest Date is {}'.format( str(len(urls)), str(utime)))
    while utime > datetime.strptime(ToDate,'%Y-%m-%d'):
        try:
            ajax = 'https://www.facebook.com/ajax/pagelet/generic.php/GroupEntstreamPagelet'
            params = {'ajaxpipe': '1',
                      'data': str({"last_view_time":0,
                                   "end_cursor":end_cursor,
                                   "group_id":ID,
                                   "multi_permalinks":[]})}
            time.sleep(1)
            resp = requests.get(ajax, params=params)
            text = re.sub(r'\\/','/', resp.text)
            
            # 紀錄加載的貼文則數，如果加載後沒有增加就跳出迴圈
            urls = list(set(urls))
            num1 = len(urls) 
            urls += re.findall('"[0-9]{1,}","url":"(https://www.facebook.com/groups.*?)/","i18n',text)
            urls = list(set(urls))
            num2 = len(urls) 
            if num2 == num1:
                break
            # 抓完網址後更新 end_cursor
            end_cursor = re.findall(',"end_cursor":"(.*?)"', resp.text)[0]
            utime = datetime.fromtimestamp(int(min(re.findall(r'utime=\\"(.*?)\\"', text))))
            print('There are {} Links in urls, and lastest Date is {}. end_cursor:{}'.format( str(len(urls)), str(utime), end_cursor))
        except:
            print(end_cursor)
            print('Error, and Try Again')
    return urls    
```

# 抓取貼文內容與摘要

```python
# 貼文區 ----
def Get_code(soup):
    for i in soup.find_all('code'):
        # 粗抓一個找貼文區的方法
        if 'abbr' in i.string:
            code = BeautifulSoup(i.string)
            break    
    return code

# 貼文顯示姓名
def Post_Name(code):
    try:
        Name = code.find('img')['aria-label']
    except:
        Name = 'ERROR'
    return Name

# 貼文ID  
def Post_ID(code):
    try:
        ID = code.find('img')['src'].split(r'.jpg')[0].split('/')[-1]
    except:
        ID = 'ERROR'    
    return ID    
    
# 貼文內容
def Post_Content(code):
    try:
        Content = code.find('div',{'data-testid':'post_message'}).text
    except:
        Content = "There's No Text Content!"
    return Content

# 時間 -----
def Post_Time(code):
    subtitle = code.find('div',{'data-testid':'story-subtitle'}).find_all('a')
    try:
        utime = subtitle[0].find('abbr')['data-utime']
        utime = datetime.fromtimestamp(int(utime))
    except:
        utime = 'ERROR'
    return utime

# 區域(可以自己寫的區域)
def Post_Loc(code):
    try:
        Loc = code.find('a',{'class':'profileLink'}).text
    except:
        Loc = 'No Infomation'
    return Loc

# 區域2(有具體連結與位置)
def Post_Loc2(code):
    subtitle = code.find('div',{'data-testid':'story-subtitle'}).find_all('a')
    try:
        Loc2 = subtitle[1].text
    except:
        Loc2 = 'No Infomation'
    return Loc2

# feedbackTargetID
def feedbackTargetID(soup):
    try:
        feedbackTargetID = re.findall('feedbackTargetID:"(.*?)"',soup.text)[0]
    except:
        feedbackTargetID = 'ERROR'
    return feedbackTargetID

# feedbackTargetID
def Get_nc_x(soup):
    try:
        _nc_x = re.findall('_nc_x=(.*?)"',soup.text)[0]
    except:
        _nc_x = 'ERROR'
    return _nc_x
```

```python
# # 貼文心情區資訊
def Get_Reaction(soup):
    for script in soup.find_all('script'):
        if 'reaction_count' in str(script):
            Reaction = script
            break
    return Reaction.string

# 貼文留言數_留言與回覆
def Post_Comments(Reaction):
    try:
        CommentCounts = re.search(r'comment_count\:\{total_count\:(.*?)\}', Reaction)[1]
    except:
        CommentCounts = str(0) 
    return CommentCounts


# 貼文留言數_純留言數()
def Post_PureComments(Reaction):
    try:
        CommentCounts = re.search(r'display_comments_count\:\{count\:(.*?)\}', Reaction)[1]
    except:
        CommentCounts = str(0) 
    return CommentCounts

# 貼文分享數
def Post_Shares(Reaction):
    try:
        ShareCounts = re.search(r'share_count\:\{count\:(.*?)\}', Reaction)[1]
    except:
        ShareCounts = str(0)
    return ShareCounts

# 心情區----
# All
def Post_Reaction_All(Reaction):
    try:
        React = re.search(r'reaction_count:\{count\:(.*?)\}', Reaction)[1]
    except:
        React = str(0)
    return React

# Like
def Post_Reaction_Like(Reaction):
    # 先給 Like 的值為 0，如果該貼文沒有資料就可以直接繼續~
    React = str(0)
    for i in re.search(r'edges\:\[(.*?)]},associated', Reaction)[1].split('},{',-1):
        if 'LIKE' in i:
            React = re.search(r'reaction_count:(.*?),node', i)[1]
    return React

# Wow
def Post_Reaction_Wow(Reaction):
    # 先給 Wow 的值為 0，如果該貼文沒有資料就可以直接繼續~
    React = str(0)
    for i in re.search(r'edges\:\[(.*?)]},associated', Reaction)[1].split('},{',-1):
        if 'WOW' in i:
            React = re.search(r'reaction_count:(.*?),node', i)[1]
    return React

# Haha
def Post_Reaction_Haha(Reaction):
    # 先給 Haha 的值為 0，如果該貼文沒有資料就可以直接繼續~
    React = str(0)
    for i in re.search(r'edges\:\[(.*?)]},associated', Reaction)[1].split('},{',-1):
        if 'HAHA' in i:
            React = re.search(r'reaction_count:(.*?),node', i)[1]
    return React

# love
def Post_Reaction_Love(Reaction):
    # 先給 Love 的值為 0，如果該貼文沒有資料就可以直接繼續~
    React = str(0)
    for i in re.search(r'edges\:\[(.*?)]},associated', Reaction)[1].split('},{',-1):
        if 'LOVE' in i:
            React = re.search(r'reaction_count:(.*?),node', i)[1]
    return React

# Angry
def Post_Reaction_Angry(Reaction):
    # 先給 Angry 的值為 0，如果該貼文沒有資料就可以直接繼續~
    React = str(0)
    for i in re.search(r'edges\:\[(.*?)]},associated', Reaction)[1].split('},{',-1):
        if 'ANGER' in i:
            React = re.search(r'reaction_count:(.*?),node', i)[1]
    return React

# Sad
def Post_Reaction_Sad(Reaction):
    # 先給 Sad 的值為 0，如果該貼文沒有資料就可以直接繼續~
    React = str(0)
    for i in re.search(r'edges\:\[(.*?)]},associated', Reaction)[1].split('},{',-1):
        if 'SORRY' in i:
            React = re.search(r'reaction_count:(.*?),node', i)[1]
    return React
```

# 組合

```python
# 文章內容與互動摘要
def PostInfo(soup):
    code = Get_code(soup)
    Reaction = Get_Reaction(soup)
    PostContent = pd.DataFrame(data = [{'Name':Post_Name(code),
                                        'ID':Post_ID(code),
                                        'Loc':Post_Loc(code),
                                        'PostTime':Post_Time(code),
                                        'Loc2':Post_Loc2(code),
                                        'Content':Post_Content(code),
                                        'Comment_counts':Post_Comments(Reaction),
                                        'PureComments_counts':Post_PureComments(Reaction),
                                        'Share_counts':Post_Shares(Reaction),
                                        'Reaction_All':Post_Reaction_All(Reaction),
                                        'Reaction_Like':Post_Reaction_Like(Reaction),
                                        'Reaction_Wow':Post_Reaction_Wow(Reaction),
                                        'Reaction_Haha':Post_Reaction_Haha(Reaction),
                                        'Reaction_Love':Post_Reaction_Love(Reaction),
                                        'Reaction_Angry':Post_Reaction_Angry(Reaction),
                                        'Reaction_Sad':Post_Reaction_Sad(Reaction),
                                        'Updatetime':datetime.now().strftime("%Y-%m-%d %H:%M"),
                                        'feedbackTargetID':feedbackTargetID(soup),
                                        'Link':resp.url}],
                            columns = ['Name', 'ID', 'Loc', 'PostTime', 'Loc2', 'Content',
                                       'Comment_counts', 'PureComments_counts', 'Share_counts',
                                       'Reaction_All','Reaction_Like', 'Reaction_Love',
                                       'Reaction_Haha','Reaction_Wow', 'Reaction_Sad',
                                       'Reaction_Angry', 'Updatetime', 'feedbackTargetID', 'Link'])
    return PostContent
```

```python
def Get_Links(url, times):
    urls = []
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)
    urls2 = [i for i in re.findall('url:"(.*?)/",',soup.text) if bool(re.search(r'^https.{1,}[0-9]{16}$', i)) if len(i)<100]
    urls += urls2
    ID = re.search('content="fb://group/(.*?)" />',resp.text)[1]
    end_cursor = re.findall('end_cursor:"(.*?)",group',resp.text)[0]
    for i in range(times):
        ajax = 'https://www.facebook.com/ajax/pagelet/generic.php/GroupEntstreamPagelet'
        data = {'ajaxpipe': '1',
                'data': str({"last_view_time":0,
                             "end_cursor":end_cursor,
                             "group_id":ID,
                             "multi_permalinks":[]})}
        time.sleep(1)
        resp = requests.get(ajax, params=data)
        # 抓完網址後更新 end_cursor
        for i in resp.text.split(',',-1):
            if re.search(r'^"end_cursor"',i):
                end_cursor = re.findall('"end_cursor":"(.*?)"', i)[0]
                break
        urls3 = re.findall(r'"https:\\/\\/www.facebook.com\\/[a-z]{1,}\\/[a-z]{1,}\\/[a-z]{1,}\\/[0-9]{1,}\\/"',resp.text)
        urls3 = [re.sub(r'\\/','/', i) for i in urls3]
        urls3 = [re.sub(r'"','', i) for i in urls3]
        urls += urls3
        print(end_cursor)
    return urls    
```
