### SELECT
select all data in the table

```python
query = '''
SELECT * 
FROM TSMC_NEWS 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### LIMIT

```python
query = '''
SELECT *
FROM TSMC_NEWS 
LIMIT 5
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### AS
rename columns

```python
query = '''
SELECT title AS NAME
    , paragraph AS ABBR
    , cateTitle AS CATEGORY
    , date as DATE
    , time AS TIME
    , url as URL
FROM TSMC_NEWS 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### ORDER BY

```python
query = '''
SELECT *
FROM TSMC_NEWS 
ORDER BY date 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

```python
query = '''
SELECT *
FROM TSMC_NEWS 
ORDER BY date DESC
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### WHERE
- =：等於
- \>:大於
- \<:小於
- <>: 不等於
- \>=: 大於等於
- <=：小於等於

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE CATETITLE = '產經'
'''
df = pd.read_sql(sql = query, con=conn)
df
```

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE date >= '2021-01-01'
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### AND

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE date >= '2021-01-01' AND  date <= '2021-07-01' 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### OR

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE date <= '2021-01-01' OR  date >= '2021-07-01' 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### BETWEEN

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE date BETWEEN '2021-01-01' AND '2021-07-01' 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### IN

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE catetitle IN ('產經', '股市', '基金') 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### NOT IN

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE catetitle NOT IN ('產經', '股市', '基金') 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### LIKE
- start with '台積電'

```python
query = '''
SELECT *
FROM TSMC_NEWS 
WHERE paragraph LIKE '台積電%' 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### GROUP BY

```python
query = '''
SELECT TIME, COUNT(*) AS cnt
FROM TSMC_NEWS 
WHERE paragraph LIKE '台積電%' 
GROUP BY date
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### HAVING

```python
query = '''
SELECT date, COUNT(*) AS cnt
FROM TSMC_NEWS 
WHERE paragraph LIKE '台積電%' 
GROUP BY date
HAVING CNT >=20
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### DISTINCT

```python
query = '''
SELECT DISTINCT date
FROM TSMC_NEWS 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### SUM

```python
query = '''
SELECT SUM(CNT)
FROM (
    SELECT date, COUNT(*) AS cnt
    FROM TSMC_NEWS 
    GROUP BY date
    ) A
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### AVG

```python
query = '''
SELECT AVG(CNT)
FROM (
    SELECT date, COUNT(*) AS cnt
    FROM TSMC_NEWS 
    GROUP BY date
    ) A
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### CASE WHEN

```python
query = '''
SELECT TITLE
		, DATE
        , CASE WHEN DATE >= '2021-01-01' THEN '2021'
         		WHEN DATE >= '2020-01-01' THEN '2020'
         		WHEN DATE >= '2019-01-01' THEN '2019'
        		ELSE '2018-' END AS year
FROM TSMC_NEWS 
'''
df = pd.read_sql(sql = query, con=conn)
df
```

### JOIN
- INNER
- LEFT

```python
query = '''
SELECT CATETITLE, COUNT(*) AS CNT
FROM TSMC_NEWS
GROUP BY CATETITLE
HAVING CNT >= 500
'''
df = pd.read_sql(sql = query, con=conn)
df
```

```python
query = '''
SELECT *
FROM TSMC_NEWS AS A
INNER JOIN (
    SELECT CATETITLE, COUNT(*) AS CNT
    FROM TSMC_NEWS
    GROUP BY CATETITLE
    HAVING CNT >= 500
    ) B
ON A.CATETITLE=B.CATETITLE
'''
df = pd.read_sql(sql = query, con=conn)
df
```
