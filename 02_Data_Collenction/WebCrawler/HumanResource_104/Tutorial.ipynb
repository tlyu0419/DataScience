# 人力銀行爬蟲案例

## 載入套件

```python
import requests
import pandas as pd
from IPython.display import clear_output
```

## 爬蟲實作

### 透過Python與伺服器建立連線

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36',
         'Referer': 'https://www.104.com.tw/'}

url= 'https://www.104.com.tw/jobs/search/list?ro=1&indcat=1003000000&area=6001001000&order=11&asc=0&page=&mode=l'
resp = requests.get(url, headers=headers)
pd.DataFrame(resp.json()['data']['list']).head(3)
```

### 透過迴圈自動生成網址與收集資料

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36',
         'Referer': 'https://www.104.com.tw/'}

df = []

for page in range(1,10):
    url= f'https://www.104.com.tw/jobs/search/list?ro=1&indcat=1003000000&area=6001001000&order=11&asc=0&page={page}&mode=l'
    print(url)
    resp = requests.get(url, headers=headers)
    ndf = pd.DataFrame(resp.json()['data']['list'])
    df.append(ndf)
    if ndf.shape[0] < 30:
        break
        
df = pd.concat(df, ignore_index=True)
df
```

### 細分查詢條件

#### 地區

```python
url = 'https://static.104.com.tw/category-tool/json/Area.json'
areacode = pd.DataFrame(requests.get(url).json()[0]['n'])
areacode = areacode.explode('n')
areacode['des2'] = areacode['n'].apply(lambda x: x['des'])
areacode['no2'] = areacode['n'].apply(lambda x: x['no'])
areacode = areacode.loc[:,['des', 'no', 'des2', 'no2']]
areacode
```

#### 產業

```python
url = 'https://static.104.com.tw/category-tool/json/Indust.json'
Indust = pd.DataFrame(requests.get(url).json())
Indust = Indust.explode('n')
Indust['des2'] = Indust['n'].apply(lambda x: x['des'])
Indust['no2'] = Indust['n'].apply(lambda x: x['no'])
Indust['n2'] = Indust['n'].apply(lambda x: x['n'])
Indust = Indust.explode('n2')
Indust['des3'] = Indust['n2'].apply(lambda x: x['des'])
Indust['no3'] = Indust['n2'].apply(lambda x: x['no'])
Indust = Indust.loc[:,['des', 'no', 'des2', 'no2', 'des3', 'no3']]
Indust
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36',
         'Referer': 'https://www.104.com.tw/'}

df = []
for area in areacode['no2'].unique():
    for indcat in Indust['no2'].unique():
        for page in range(1,101):
            url = f'https://www.104.com.tw/jobs/search/list?ro=1&indcat={indcat}&area={area}&order=11&asc=0&page={page}&mode=l'
            print(url)
            resp = requests.get(url, headers=headers)
            ndf = pd.DataFrame(resp.json()['data']['list'])
            df.append(ndf)
            if ndf.shape[0] < 30:
                break
        clear_output()
df = pd.concat(df, ignore_index=True)
df
```

### 容錯機制

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36',
         'Referer': 'https://www.104.com.tw/'}

df = []
for area in areacode['no2'].unique():
    for indcat in Indust['no2'].unique():
        page = 1
        while page <= 100:
            try:
                url = f'https://www.104.com.tw/jobs/search/list?ro=1&indcat={indcat}&area={area}&order=11&asc=0&page={page}&mode=l'
                print(url)
                resp = requests.get(url, headers=headers)
                ndf = pd.DataFrame(resp.json()['data']['list'])
                df.append(ndf)
                if ndf.shape[0] < 30:
                    break
                page = page + 1
            except:
                print('==================== Error and retry ====================')
        clear_output()
df = pd.concat(df, ignore_index=True)
df
```
