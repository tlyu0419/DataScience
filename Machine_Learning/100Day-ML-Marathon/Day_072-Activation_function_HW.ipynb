```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

# Sigmoid
#sigmoid = lambda x: 1 / (1 + np.exp(-x))

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def dsigmoid(x):
    return (x * (1 - x))

# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
x = plt.linspace(-10, 10, 100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, sigmoid(x), 'b', label='linspace(-10, 10, 10)')

# Draw the grid line in background
plt.grid()

# title
plt.title('Sigmoid Function')

# write the Sigmoid formula
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{e^{-x}}$', fontsize=15)

# resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

# create the grath
plt.show
```

# softmax

```python
def softmax(x):
    return np.exp(x) / float(sum(np.exp(x)))

# x = np.arange(0, 1.0, 0.01)
x = plt.linspace(-5, 5, 100)

# resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))

print(softmax(x))
plt.plot(x, softmax(x), 'r')
plt.show()
```

# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
def relu(z):
  return (np.abs(z) + z) / 2
  
def relu_prime(z):
  return np.where(z > 0, 1, 0)

z = plt.linspace(-10, 10, 100)
```

```python
# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(z, relu(z), 'b', label='linspace(-10, 10, 0.1)')

# Draw the grid line in background
plt.grid()

# title
plt.title('Relu Function')

# write the Sigmoid formula

# resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

# create the grath
plt.show
```

```python
# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(z, relu_prime(z), 'b', label='linspace(-10, 10, 0.1)')

# Draw the grid line in background
plt.grid()

# title
plt.title('dRelu Function')

# write the Sigmoid formula

# resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

# create the grath
plt.show
```
