# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print(f'Path of read in data: {f_app}')
app_train = pd.read_csv(f_app)
app_train.head()
```

請參考 HomeCredit_columns_description.csv 的欄位說明，
觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
app_train.dtypes.value_counts()
```

```python
# 在此僅先針對數值型欄位進行異常值偵測
# 先篩選數值型的欄位
dtype_select_num = app_train.select_dtypes(include=[np.number])
dtype_select_num.dtypes.value_counts()
```

```python
# 如果攔位下只有2個類別(Dummy variable)，也暫先排除
for i in dtype_select_num:                        #run loop
    if len(np.unique(dtype_select_num[i])) == 2:  #若只有2類
        del dtype_select_num[i]                   #刪除        
dtype_select_num.dtypes.value_counts()
```

```python
# 繪製箱型圖，檢視各變量是否有異常值。
# 由於變量相當多，在此將多張圖整併後一次輸出
ncol = 4 # 每列幾張圖
nrow = int(np.ceil(dtype_select_num.columns.size/ncol)) # 根據前面的張數，決定需要幾個列
fig = plt.subplots(nrow, ncol, figsize=(15,20))
for i, feature in enumerate(dtype_select_num.columns):
    plt.subplot(nrow, ncol, i+1)
    sns.boxplot(x=feature, data = dtype_select_num)
    plt.tight_layout()
```

```python
# 透過以上的箱型圖，我們可以輕易發現以上許多欄位都包含異常值。

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# DAYS_EMPLOYED
# ...
```

```python
# 最大值離平均與中位數很遠
app_train['AMT_INCOME_TOTAL'].describe()
```

```python
# 繪製 Empirical Cumulative Density Plot (ECDF)
def ECDF(data):   
    F = data.value_counts().sort_index().cumsum()
    m = len(data)
    return F / m

cdf = ECDF(app_train['AMT_INCOME_TOTAL'])
cdf
```

```python
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
# plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
# plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()
```

```python
# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

## 補充：Normal dist 的 ECDF

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = ECDF(app_train['REGION_POPULATION_RELATIVE'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)

cdf = ECDF(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])


plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min() * 0.95, cdf.index.max() * 1.05])
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
