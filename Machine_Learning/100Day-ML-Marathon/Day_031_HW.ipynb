# 品質檢測
各大公司的客服中心每天都會接聽相當大量的客戶電話，包含售前、售後的服務。
公司為了避免收到政府機關、民間協會或客戶直接的投訴，會另外再設立一個「品質管理」的部門。為了要協助品管部門監測客服人員的服務品質，會再要求服務人員在結束服務後，用文字記錄該次服務的服務內容。品管的流程大致如下：
「客戶來電」→「人員摘要紀錄服務內容」→「品管檢核文字記錄」→「找出服務缺失的案件」。

但是這中間經過了三次的資料轉換流程，每次的轉換不只沒有一致的標準，甚至其中還會有惡意的欺騙行為發生。如人員為了規避品管發現服務瑕疵，簡化或竄改客戶的投訴電話。顯然有大幅的改善空間。如果我們直接將客戶的聲音資料轉換為電腦可以理解的資料結構，並藉以訓練出一個監督式的模型偵測客戶的投訴行為。將可以大幅提升客服人員的作業效率、避免後續的管理成本。為公司提升作業效率並大幅節省後勤的作業成本。
