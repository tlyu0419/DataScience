# Pytorch 實作邏輯回歸

```python
import torch
```

```python
n_data = torch.ones(100, 2)
xy0 = torch.normal(2 * n_data, 1.5)
c0 = torch.zeros(100)
xy1 = torch.normal(-2 *n_data, 1.5)
c1 = torch.ones(100)

x,y = torch.cat((xy0, xy1),0).split(1, dim=1)
x = x.squeeze()
y = y.squeeze()
c = torch.cat((c0, c1), 0)
```

## 視覺化

```python
import matplotlib.markers as mmarkers
import matplotlib.pyplot as plt
def plot(x, y, c):
    ax = plt.gca()
    sc = ax.scatter(x, y, color='black')
    paths = []
    for i in range(len(x)):
        if c[i].item()==0:
            marker_obj = mmarkers.MarkerStyle('o')
        else:
            marker_obj = mmarkers.MarkerStyle('x')
        path = marker_obj.get_path().transformed(marker_obj.get_transform())
        paths.append(path)
    sc.set_paths(paths)
```

```python
plot(x, y, c)
```

## 定義模型

```python
w1 = torch.tensor([1.,], requires_grad=True)
w2 = torch.tensor([1.,], requires_grad=True)
b = torch.zeros((1), requires_grad=True)
loss = ((torch.sigmoid(x*w1 + w2*y + b) - c)**2).mean()
loss
```

```python
x*w1 + y*w2 + b
```

## 訓練模型
- 目前有兩個 weight, 要在平面上劃線不太容易
- 未來換成線性回歸會容易些
- 也許作者當初是考量到要畫在平面上不容易，所以才拿掉 y 的 weight

```python
xx = torch.arange(-4, 5)
lr = 0.02

for iteration in range(1000):
    # 前向傳播
    loss = ((torch.sigmoid(x*w1 + y*w2 + b)-c)**2).mean()

    # 反向傳播
    loss.backward()
    
    # 更新參數
    b.data.sub_(lr*b.grad) # b = b - lr*b.grad
    w1.data.sub_(lr*w1.grad) # w = w - lr*w.grad
    w2.data.sub_(lr*w2.grad) # w2 = w2 - lr*w.grad
    
    # 繪圖
    if iteration % 3 == 0:
        plot(x, y, c)
        yy = w1*xx+b
        plt.plot(xx.data.numpy(), yy.data.numpy(), 'r-', lw=5)
        plt.text(-4, 2, 'Loss=%.4f'%loss.data.numpy(), fontdict={'size':20, 'color':'black'})
        plt.xlim(-4, 4)
        plt.ylim(-4, 4)
        plt.title(f'Iteration:{iteration}\nw:{w1.data.numpy()},b:{b.data.numpy()}')
        plt.show(0.5)
        if loss.data.numpy() < 0.03:
            break
```
